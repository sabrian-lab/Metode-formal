## Dasar dari Banyak Metode Formal (dan Sebagian Besar Matematika Modern)

**Definisi:** Himpunan adalah koleksi objek yang terdefinisi dengan baik yang dianggap sebagai satu kesatuan. Objek-objek dalam himpunan disebut elemen. Himpunan secara konvensional dilambangkan dengan huruf kapital, seperti A, B, C, dll. Dua himpunan A dan B dikatakan sama, ditulis A = B, jika keduanya memiliki elemen yang sama.

Kita menulis:
- x $\in$ A, jika x adalah elemen dari A
- x $\notin$ A, jika x bukan elemen dari A

**Definisi:**
- **A adalah subset dari B (A $\subseteq$ B )**, jika untuk semua x $\in$ A : x $\in$ B
- **A adalah proper subset dari B (A $\subset$ B)**, jika A $\subseteq$ B dan A $\neq$ B

```python
import sympy
from sympy import Symbol, FiniteSet, EmptySet, UniversalSet
from IPython.display import display

# Define symbols for elements
x, y, z = Symbol('x'), Symbol('y'), Symbol('z')

display('--- Set Definition and Elements ---')
# Define sets
A = FiniteSet(1, 2, 3)
B = FiniteSet(2, 3, 4)
C = FiniteSet(1, 2, 3, 4, 5)
D = FiniteSet(1, 2, 3) # A and D have the same elements
E = EmptySet

display("A:",A,"B:",B,"C:",C,"D:",D,"E:",E)


# Check if an element is in a set
display(f"Is 1 an element of A? {1 in A}")
display(f"Is 4 an element of A? {4 in A}")

# Check if an element is not in a set
display(f"Is 5 not an element of A? {5 not in A}")
display(f"Is 2 not an element of A? {2 not in A}")

# Check for set equality

display(f"Is A equal to B? {A == B}")
display(f"Is A equal to D? {A == D}")

display('\n--- Subset and Proper Subset ---')
# Check for subset relation (A is a subset of B if all elements of A are in B)
display(f"Is A a subset of C? {A.is_subset(C)}") # {1,2,3} is subset of {1,2,3,4,5}
display(f"Is B a subset of A? {B.is_subset(A)}") # {2,3,4} is not subset of {1,2,3}
display(f"Is A a subset of A? {A.is_subset(A)}") # A is always a subset of itself
display(f"Is EmptySet() a subset of A? {E.is_subset(A)}") # Empty set is a subset of every set

# Check for proper subset relation (A is a proper subset of B if A is a subset of B and A != B)
display(f"Is A a proper subset of C? {A.is_proper_subset(C)}") # True, because A subset C and A != C
display(f"Is A a proper subset of A? {A.is_proper_subset(A)}") # False, because A == A
```

## Himpunan Kosong dan Himpunan Bilangan

$$\emptyset = \{\}$$, tidak mengandung elemen apa pun

**Himpunan bilangan:**

## Operasi Himpunan (Set Operations) ##
Misalkan $A$, $B$, dan $C$ adalah himpunan

- $A \cup B = \{x \mid x \in A \text{ atau } x \in B\}$ adalah **union** (gabungan) dari $A$ dan $B$
  - $A \cup B = B \cup A$ ($\cup$ bersifat komutatif)
  - $A \cup (B \cup C ) = (A \cup B ) \cup C$ ($\cup$ bersifat asosiatif)
  - $A \cup A = A$ ($\cup$ bersifat idempoten)
  - Lebih umum: $B \subseteq A$ menyiratkan $A \cup B = A$

- $A \cap B = \{x \mid x \in A \text{ dan } x \in B\}$ adalah **intersection** (irisan) dari $A$ dan $B$
  - $\cap$ bersifat asosiatif, komutatif, idempoten
  - $B \subseteq A$ menyiratkan $A \cap B = B$

- $A - B = \{x \mid x \in A, x \notin B\}$ adalah **(selisih himpunan)** dari $A$ dan $B$
  - $A - A = \emptyset$
  - $A - \emptyset = A$

- $C = A \uplus B$ disebut **disjoint union** (gabungan disjoin) dari $A$ dan $B$, jika $C = A \cup B$ dan $A \cap B = \emptyset$

```python
# Define example sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
C = {5, 6, 7, 8}
D = {1, 2}
E = {7, 8, 9} # For disjoint union example

print("--- Set Operations ---")
print(f"Set A: {A}")
print(f"Set B: {B}")
print(f"Set C: {C}")
print(f"Set D: {D}")
print(f"Set E: {E}")

# Union (A ∪ B)
union_AB = A.union(B)
print(f"\nUnion (A \u222a B): {union_AB}")

# Properties of Union
print(f"Union (B \u222a A) (commutative): {B.union(A)}")
print(f"Union (A \u222a A) (idempotent): {A.union(A)}")
print(f"A \u222a (B \u222a C) (associative): {A.union(B.union(C))}")
print(f"(A \u222a B) \u222a C (associative): {(A.union(B)).union(C)}")
print(f"B \u2286 A implies A \u222a B = A (D \u2286 A): {A.union(D)}")

# Intersection (A ∩ B)
intersection_AB = A.intersection(B)
print(f"\nIntersection (A \u2229 B): {intersection_AB}")

# Properties of Intersection
print(f"Intersection (B \u2229 A) (commutative): {B.intersection(A)}")
print(f"Intersection (A \u2229 A) (idempotent): {A.intersection(A)}")
print(f"A \u2229 (B \u2229 C) (associative): {A.intersection(B.intersection(C))}")
print(f"(A \u2229 B) \u2229 C (associative): {(A.intersection(B)).intersection(C)}")
print(f"B \u2286 A implies A \u2229 B = B (D \u2286 A): {A.intersection(D)}")

# Difference (A - B)
difference_AB = A.difference(B)
print(f"\nDifference (A - B): {difference_AB}")
print(f"Difference (B - A): {B.difference(A)}")

# Properties of Difference
print(f"A - A = \u2205: {A.difference(A)}")
print(f"A - \u2205 = A: {A.difference(set())}")

# Disjoint Union (conceptual in Python - no direct operator, check for disjoint first)
# Sets are disjoint if their intersection is an empty set.
print(f"\nIs A and E disjoint? {A.isdisjoint(E)}") # True, as A \u2229 E = \u2205
print(f"Is A and B disjoint? {A.isdisjoint(B)}") # False, as A \u2229 B = {3, 4}

# If sets are disjoint, their 'disjoint union' is simply their union.
disjoint_union_AE = A.union(E) if A.isdisjoint(E) else None
print(f"Disjoint Union (A \u228e E) (if disjoint): {disjoint_union_AE}")
```

## Himpunan Kuasa (Power Set)
**Definisi:** Misalkan $A$ adalah sebuah himpunan. Himpunan kuasa $\mathbb{P}A$ adalah himpunan dari semua subset dari $A$, yaitu $\mathbb{P}A = \{B \mid B \subseteq A\}$

$\mathbb{P}_1 A$ adalah himpunan subset $A$ yang tidak kosong

**Contoh:**
- $\mathbb{P}\emptyset = \{\emptyset\}$
- $\mathbb{P}\{1, 2, 3\} = \{\emptyset, \{1\}, \{2\}, \{3\}, \{1, 2\}, \{1, 3\}, \{2, 3\}, \{1, 2, 3\}\}$
- $\mathbb{P}_1\{1, 2\} = \{\{1\}, \{2\}, \{1, 2\}\}$

**Peringatan:** Mendaftarkan himpunan kuasa yang besar dapat sangat rumit!

## Produk Kartesian dan Tupples

**Definisi:** Misalkan $A$, $B$ adalah himpunan. $A \times B = \{(a, b) \mid a \in A, b \in B\}$ disebut **produk Kartesian (biner)** dari $A$ dan $B$.

Secara lebih umum, misalkan $A_1, \dots, A_n$ adalah himpunan. Maka $A_1 \times \dots \times A_n = \{(a_1,\dots, a_n \mid a_i \in A_i \text{ untuk } 1 \le i \le n\}$ adalah produk Kartesian dari $A_1, \dots, A_n$.

Elemen-elemen produk Kartesian biner disebut **tupel (biner)**.
Elemen-elemen produk Kartesian dari $n$ himpunan disebut **$n$-tupel**.

**Contoh:**
- $\{1, 2, 3\} \times \{a, b\} = \{(1, a), (1, b), (2, a), (2, b), (3, a), (3, b)\}$
- $\mathbb{N} \times \emptyset = \emptyset$


```python
from itertools import product

print("--- Cartesian Product Examples ---")

# Example 1: {1, 2, 3} x {a, b}
set_A = {1, 2, 3}
set_B = {'a', 'b'}

# Calculate Cartesian product
cartesian_product_AB = set(product(set_A, set_B))

print(f"Set A: {set_A}")
print(f"Set B: {set_B}")
print(f"A x B: {cartesian_product_AB}")

# Example 2: N x empty_set (Conceptual representation)
# In set theory, any set multiplied by the empty set results in the empty set.
# Python's itertools.product will also return an empty iterator if any input is empty.

set_N_conceptual = {0, 1, 2} # Representing a finite subset of Natural Numbers
empty_set = set()

cartesian_product_N_empty = set(product(set_N_conceptual, empty_set))

print(f"\nSet N (conceptual subset): {set_N_conceptual}")
print(f"Empty Set: {empty_set}")
print(f"N x Empty Set: {cartesian_product_N_empty}")

# Example 3: Cartesian product of three sets (A x B x C)
set_C = {True, False}
cartesian_product_ABC = set(product(set_A, set_B, set_C))

print(f"\nSet A: {set_A}")
print(f"Set B: {set_B}")
print(f"Set C: {set_C}")
print(f"A x B x C: {cartesian_product_ABC}")

# The elements are tuples. For n sets, the elements are n-tuples.
print(f"\nType of an element in A x B: {type(next(iter(cartesian_product_AB)))}")
print(f"Type of an element in A x B x C: {type(next(iter(cartesian_product_ABC)))}")
```

## Relasi ##
**Definisi:** Misalkan $A$, $B$ adalah himpunan. $R \in \mathbb{P}(A \times B)$ disebut relasi biner atas $A$ dan $B$.

Kasus yang sering terjadi: $A = B$ (relasi homogen)

Dalam notasi Z, kita juga menulis $A \leftrightarrow B$ daripada $\mathbb{P}(A \times B)$ untuk himpunan semua relasi atas $A$ dan $B$.
Demikian pula, kita menulis $R: A \leftrightarrow B$ daripada $R \in \mathbb{P}(A \times B)$
Kita sering menulis $R(a, b)$ untuk $(a, b) \in R$
Banyak relasi khusus ditulis dalam bentuk infix ($a > b$, $a = b$, ...)

**Definisi:** Jika $R$ adalah relasi, $R^{-1} = \{(b, a) \mid R(a, b)\}$ disebut relasi invers dari $R$.

**Contoh:**
- $>$ (atas $\mathbb{N}$) adalah relasi atas $\mathbb{N}$, $\mathbb{N}$. $(4, 1) \in \text{>}$ (lebih familiar: $4 > 1$)
- $ >^{-1} $ adalah $ <$
- $\{(a, a) \mid a \in A\} \in A \leftrightarrow A$ adalah relasi kesetaraan pada $A$ $(=)$, dan $=^{-1} = =$


```python
# Define example sets A and B
A = {1, 2, 3}
B = {'x', 'y', 'z'}

print("--- Binary Relations Examples ---")
print(f"Set A: {A}")
print(f"Set B: {B}")

# Example 1: A binary relation R from A to B (R subset of A x B)
# R = {(1, 'x'), (2, 'y'), (3, 'x')}
R1 = {(1, 'x'), (2, 'y'), (3, 'x')}
print(f"\nRelation R1 (A to B): {R1}")

# Check if an element is in the relation
print(f"Is (1, 'x') in R1? {(1, 'x') in R1}")
print(f"Is (1, 'y') in R1? {(1, 'y') in R1}")

# Example 2: A homogenous relation R on A (R subset of A x A)
# Let's define R2 as 'less than' relation on a subset of N
set_N_subset = {1, 2, 3, 4}
R2 = {(a, b) for a in set_N_subset for b in set_N_subset if a < b}
print(f"\nSet N (subset for example): {set_N_subset}")
print(f"Relation R2 (less than '<' on N subset): {R2}")

# Check using infix notation (conceptually)
print(f"Is (2, 4) in R2 (i.e., 2 < 4)? {(2, 4) in R2}")

# --- Inverse Relation ---
print("\n--- Inverse Relation Examples ---")

# Calculate the inverse of R1
R1_inverse = {(b, a) for a, b in R1}
print(f"R1: {R1}")
print(f"Inverse of R1 (R1^-1): {R1_inverse}")

# Calculate the inverse of R2
R2_inverse = {(b, a) for a, b in R2}
print(f"R2 (less than): {R2}")
print(f"Inverse of R2 (R2^-1, which is 'greater than'): {R2_inverse}")

# Example 3: Equality relation on A and its inverse
set_A_equal = {'apple', 'banana', 'cherry'}
Equality_A = {(x, x) for x in set_A_equal}
print(f"\nSet A for equality: {set_A_equal}")
print(f"Equality Relation on A: {Equality_A}")

Equality_A_inverse = {(b, a) for a, b in Equality_A}
print(f"Inverse of Equality Relation on A: {Equality_A_inverse}")
print(f"Is Equality Relation equal to its inverse? {Equality_A == Equality_A_inverse}")
```

## Properti Relasi
**Definisi:** Misalkan $R \in A \leftrightarrow A$ adalah relasi homogen atas $A$.

$R$ disebut **refleksif**, jika $R(a, a)$ untuk semua $a \in A$.
$R$ disebut **simetris**, jika $R(a, b)$ menyiratkan $R(b, a)$ untuk semua $a, b \in A$.
$R$ disebut **transitif**, jika $R(a, b)$, $R(b, c)$ menyiratkan $R(a, c)$ untuk semua $a, b, c \in A$.
Sebuah relasi yang refleksif, simetris, dan transitif disebut **relasi ekuivalen**.

Terdapat dua versi antisimetri yang berbeda, tetapi saling terkait:
1.  $R$ disebut **(secara ketat) antisimetris**, jika $R(a, b)$ menyiratkan $(b, a) \notin R$ untuk semua $a, b \in A$.
2.  $R$ disebut **(secara lemah) antisimetris**, jika untuk semua $a, b \in A : R(a, b)$ dan $R(b, a)$ menyiratkan $a = b$.

$R$ disebut **antirefleksif**, jika tidak ada $a \in A : R(a, a)$.

```python
from itertools import product

# Helper function to generate all possible pairs for A x A
def get_cartesian_product_AxA(A):
    return set(product(A, A))

# --- Property Checking Functions ---

def is_reflexive(R, A):
    """Checks if a relation R is reflexive on set A.
    R is reflexive if (a, a) is in R for all a in A.
    """
    for a in A:
        if (a, a) not in R:
            return False
    return True

def is_symmetric(R):
    """Checks if a relation R is symmetric.
    R is symmetric if R(a, b) implies R(b, a).
    """
    for a, b in R:
        if (b, a) not in R:
            return False
    return True

def is_transitive(R):
    """Checks if a relation R is transitive.
    R is transitive if R(a, b) and R(b, c) implies R(a, c).
    """
    for a, b in R:
        for c, d in R:
            if b == c and (a, d) not in R:
                return False
    return True

def is_strictly_antisymmetric(R, A):
    """Checks if a relation R is strictly antisymmetric (strong antisymmetry).
    R is strictly antisymmetric if R(a, b) implies (b, a) not in R for all a, b in A.
    This implies R must also be antireflexive.
    """
    # A strictly antisymmetric relation cannot contain any (a,a) pairs.
    # So, first check for antireflexivity.
    if not is_antireflexive(R, A):
        return False

    for a, b in R:
        if a != b and (b, a) in R:
            return False
    return True

def is_weakly_antisymmetric(R):
    """Checks if a relation R is weakly antisymmetric.
    R is weakly antisymmetric if for all a, b: R(a, b) and R(b, a) implies a = b.
    """
    for a, b in R:
        if (b, a) in R and a != b:
            return False
    return True

def is_antireflexive(R, A):
    """Checks if a relation R is antireflexive on set A.
    R is antireflexive if for no a in A: R(a, a).
    """
    for a in A:
        if (a, a) in R:
            return False
    return True


# --- Example Usage ---

# Define a base set
A = {1, 2, 3}
print(f"Base Set A: {A}\n")

# Example Relations (subsets of A x A)
R_empty = set() # Empty relation
R_eq = {(1, 1), (2, 2), (3, 3)} # Equality relation
R_lt = {(1, 2), (1, 3), (2, 3)} # Less than relation
R_le = {(1, 1), (2, 2), (3, 3), (1, 2), (1, 3), (2, 3)} # Less than or equal to relation
R_custom1 = {(1, 2), (2, 1), (1, 1)} # Not transitive, not antisymmetric
R_custom2 = {(1, 1), (2, 2), (3, 3), (1, 2), (2, 1)} # Reflexive, symmetric, not transitive

relations = {
    "R_empty": R_empty,
    "R_eq": R_eq,
    "R_lt": R_lt,
    "R_le": R_le,
    "R_custom1": R_custom1,
    "R_custom2": R_custom2
}

for name, R in relations.items():
    print(f"--- Analyzing Relation: {name} = {R} ---")
    print(f"  Reflexive: {is_reflexive(R, A)}")
    print(f"  Symmetric: {is_symmetric(R)}")
    print(f"  Transitive: {is_transitive(R)}")
    print(f"  Strictly Antisymmetric: {is_strictly_antisymmetric(R, A)}")
    print(f"  Weakly Antisymmetric: {is_weakly_antisymmetric(R)}")
    print(f"  Antireflexive: {is_antireflexive(R, A)}")

    # Check for Partial Orderings
    is_spo = is_strictly_antisymmetric(R, A) and is_transitive(R)
    print(f"  Is Strict Partial Ordering (Definition 1)? {is_spo}")

    is_po = is_reflexive(R, A) and is_weakly_antisymmetric(R) and is_transitive(R)
    print(f"  Is Partial Ordering (Definition 2)? {is_po}")

    # Check for Equivalence Relation
    is_eq_rel = is_reflexive(R, A) and is_symmetric(R) and is_transitive(R)
    print(f"  Is Equivalence Relation? {is_eq_rel}\n")
```

## Urutan Parsial (Partial Orderings)
**Definisi (1):** Sebuah relasi $> : A \leftrightarrow A$ disebut **urutan parsial (ketat)** pada $A$ jika:
- $>$ adalah (secara kuat) antisimetris
- $>$ adalah transitif
- ($>$ adalah antirefleksif) (berasal dari antisimetri (kuat))

**Definisi (2):** Sebuah relasi $\ge : A \leftrightarrow A$ disebut **urutan parsial** pada $A$ jika:
- $\ge$ adalah (secara lemah) antisimetris
- $\ge$ adalah transitif
- $\ge$ adalah refleksif

($\ge - =$) disebut bagian ketat dari $\ge$ dan merupakan urutan parsial dalam pengertian pertama
$> \cup =$ adalah urutan parsial dalam pengertian kedua.

```python
import collections

# --- Helper Functions for Function Properties ---

def is_partial_function(f):
    """Checks if a relation f is a partial function.
    A relation f is a partial function if for every (a, b) in f, there's no (a, c) in f where b != c.
    """
    domain_elements = collections.defaultdict(list)
    for a, b in f:
        domain_elements[a].append(b)
    for a in domain_elements:
        if len(domain_elements[a]) > 1:
            return False # Found an 'a' mapping to multiple 'b's
    return True

def get_domain(f):
    """Calculates the domain of a relation/function.
    The domain is the set of all first elements of the pairs in f.
    """
    return {a for a, b in f}

def get_range(f):
    """Calculates the range of a relation/function.
    The range is the set of all second elements of the pairs in f.
    """
    return {b for a, b in f}

def is_total_function(f, A):
    """Checks if a partial function f is a total function on set A.
    A total function's domain must be equal to set A.
    """
    if not is_partial_function(f):
        return False
    return get_domain(f) == A

def is_injective(f):
    """Checks if a partial function f is injective (one-to-one).
    Injective if for all (a, c) in f and (b, c) in f, implies a = b.
    """
    if not is_partial_function(f):
        return False
    range_elements = collections.defaultdict(list)
    for a, b in f:
        range_elements[b].append(a)
    for b in range_elements:
        if len(range_elements[b]) > 1:
            return False # Found a 'b' mapped by multiple 'a's
    return True

def is_surjective(f, B):
    """Checks if a partial function f is surjective (onto) to set B.
    Surjective if the range of f is equal to set B.
    """
    if not is_partial_function(f):
        return False
    return get_range(f) == B

def is_bijective(f, A, B):
    """Checks if a function f is bijective (one-to-one correspondence) between A and B.
    Bijective if it is a total function, injective, and surjective.
    """
    return is_total_function(f, A) and is_injective(f) and is_surjective(f, B)

# --- Example Usage ---

# Define sets A and B
A_set = {1, 2, 3}
B_set = {'x', 'y', 'z'}
C_set = {'p', 'q', 'r', 's'}

print(f"Set A: {A_set}")
print(f"Set B: {B_set}")
print(f"Set C: {C_set}\n")

# Example 1: A partial function
f1 = {(1, 'x'), (2, 'y')}
print(f"--- Analyzing f1 = {f1} ---")
print(f"  Is Partial Function? {is_partial_function(f1)}")
print(f"  Domain: {get_domain(f1)}")
print(f"  Range: {get_range(f1)}")
print(f"  Is Total Function (A to B)? {is_total_function(f1, A_set)}")
print(f"  Is Injective? {is_injective(f1)}")
print(f"  Is Surjective (to B)? {is_surjective(f1, B_set)}")
print(f"  Is Bijective (A to B)? {is_bijective(f1, A_set, B_set)}\n")

# Example 2: A non-function (one input maps to multiple outputs)
f2 = {(1, 'x'), (1, 'y'), (2, 'z')}
print(f"--- Analyzing f2 = {f2} ---")
print(f"  Is Partial Function? {is_partial_function(f2)}\n")

# Example 3: A total and bijective function
f3 = {(1, 'x'), (2, 'y'), (3, 'z')}
print(f"--- Analyzing f3 = {f3} ---")
print(f"  Is Partial Function? {is_partial_function(f3)}")
print(f"  Domain: {get_domain(f3)}")
print(f"  Range: {get_range(f3)}")
print(f"  Is Total Function (A to B)? {is_total_function(f3, A_set)}")
print(f"  Is Injective? {is_injective(f3)}")
print(f"  Is Surjective (to B)? {is_surjective(f3, B_set)}")
print(f"  Is Bijective (A to B)? {is_bijective(f3, A_set, B_set)}\n")

# Example 4: A total but not injective (many-to-one) function
f4 = {(1, 'x'), (2, 'x'), (3, 'y')}
print(f"--- Analyzing f4 = {f4} ---")
print(f"  Is Partial Function? {is_partial_function(f4)}")
print(f"  Domain: {get_domain(f4)}")
print(f"  Range: {get_range(f4)}")
print(f"  Is Total Function (A to B)? {is_total_function(f4, A_set)}")
print(f"  Is Injective? {is_injective(f4)}")
print(f"  Is Surjective (to B)? {is_surjective(f4, B_set)}")
print(f"  Is Bijective (A to B)? {is_bijective(f4, A_set, B_set)}\n")

# Example 5: A total and injective but not surjective function (A to C)
f5 = {(1, 'p'), (2, 'q'), (3, 'r')}
print(f"--- Analyzing f5 = {f5} ---")
print(f"  Is Partial Function? {is_partial_function(f5)}")
print(f"  Domain: {get_domain(f5)}")
print(f"  Range: {get_range(f5)}")
print(f"  Is Total Function (A to C)? {is_total_function(f5, A_set)}")
print(f"  Is Injective? {is_injective(f5)}")
print(f"  Is Surjective (to C)? {is_surjective(f5, C_set)}") # Note: C_set is larger
print(f"  Is Bijective (A to C)? {is_bijective(f5, A_set, C_set)}\n")
```

## Fungsi (Functions)
**Definisi:** Misalkan $A$, $B$ adalah himpunan dan $f : A \leftrightarrow B$ adalah relasi.

$f$ disebut **fungsi (parsial)** jika dan hanya jika $(a, b) \in f$ dan $(a, c) \in f$ menyiratkan $b = c$. Dalam kasus tersebut, kita menulis $f : A \longrightarrow B$ dan $f(a) = b$.

$\text{dom } f = \{a \mid (a, b) \in f \}$ disebut **domain** dari $f$
$\text{ran } f = \{b \mid (a, b) \in f \}$ disebut **range** dari $f$

Jika $\text{dom } f = A$, maka $f$ disebut **fungsi total**, dan kita menulis $f : A \longrightarrow B$
Jika $\text{ran } f = B$, maka $f$ disebut **surjektif**, dan kita menulis $f : A \twoheadrightarrow B$ atau $f : A \twoheadrightarrow B$ (jika $f$ adalah total)
Jika $f (a) = c$ dan $f (b) = c$ menyiratkan $a = b$, maka $f$ disebut **injektif**
Jika $f$ bersifat injektif dan surjektif, maka ia disebut **bijektif** atau **pemetaan satu-ke-satu**
